#length of cycle


```python
#lengthofcycle

y1=data["LengthofCycle"]
df1=data.drop("LengthofCycle",axis=1)
```

```python
# create a dictionary 'models' with various models
models = {}

# Logistic Regression
from sklearn.linear_model import LogisticRegression
models['Logistic Regression'] = LogisticRegression()

# Support Vector Machines
from sklearn.svm import LinearSVC
models['Support Vector Machines'] = LinearSVC()

# Decision Trees
from sklearn.tree import DecisionTreeClassifier
models['Decision Trees'] = DecisionTreeClassifier()

# Random Forest
from sklearn.ensemble import RandomForestClassifier
models['Random Forest'] = RandomForestClassifier()

# Naive Bayes
from sklearn.naive_bayes import GaussianNB
models['Naive Bayes'] = GaussianNB()

# K-Nearest Neighbors
from sklearn.neighbors import KNeighborsClassifier
models['K-Nearest Neighbor'] = KNeighborsClassifier()
```

```python
# splitting data into training and testing set
from sklearn.model_selection import train_test_split
X_train1, X_test1, y_train1, y_test1 = train_test_split(df1, y1 , test_size=0.25, random_state=0)
```

```python

from sklearn.metrics import accuracy_score, precision_score, recall_score

accuracy1, precision1, recall1 = {}, {}, {}
```

```python
# fitting models and updating their accuracy metrics to models dictionary
for key in models.keys():

    # Fit the classifier
    models[key].fit(X_train1, y_train1)

    # Make predictions
    predictions = models[key].predict(X_test1)

    # Calculate metrics
    accuracy1[key] = accuracy_score(predictions, y_test1)
    precision1[key] = precision_score(predictions, y_test1,average='weighted')
    recall1[key] = recall_score(predictions, y_test1,average='weighted')
```

```python
df_model1 = pd.DataFrame(index=models.keys(), columns=['Accuracy', 'Precision', 'Recall'])
df_model1['Accuracy'] = accuracy1.values()
df_model1['Precision'] = precision1.values()
df_model1['Recall'] = recall1.values()


df_model1
```

```python
# final model
model=models["Random Forest"]
```

```python
import pickle
```

```python
# save model as binary file
with open('model_cycle_len','wb') as f:
  pickle.dump(model,f)
```
